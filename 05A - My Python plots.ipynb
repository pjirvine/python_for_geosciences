```python
from netCDF4 import Dataset
import matplotlib.pylab as plt
import numpy as np

%matplotlib inline
```

# Example usage - Histograms of anomalies

First load up data for GeoMIP experiments and check it looks right.

```python
# GeoMIP experiments
experiments = ['G1','abrupt4xCO2','piControl']

# Load NetCDF files for precip for NorESM
netcdf_files = [Dataset('pr_Amon_NorESM1-M_' + X + '_r1i1p1_11-50_ann_mean.nc') for X in experiments]

# Extract precipitation variable
precip_variable = [X['pr'] for X in netcdf_files]

# Plot data for piControl
plt.imshow(precip_variable[0][0,:,:]);
```

Next convert to numpy arrays and convert from mm/s to mm/day

```python
# Convert to numpy array and convert from mm/s to mm/day
precip_array = [np.asarray(X) * 86400 for X in precip_variable]

print precip_array[0].shape
```

Now flatten arrays so they can be used in histogram plots.

```python
precip_flat = [X.flatten() for X in precip_array]
```

Then take anomalies

```python
CO2_control_anom = precip_flat[1] - precip_flat[0]
G1_control_anom = precip_flat[2] - precip_flat[0]
G1_CO2_anom = precip_flat[2] - precip_flat[1]
```

Plot the anomaly in mm/day

```python
X = plt.hist([CO2_control_anom, G1_control_anom], 100, histtype='step', color = ['r','b'], range=[-2,2]);
```

Now do the same for % change

```python
CO2_control_pc_anom = 100.0 * (precip_flat[1] / precip_flat[0] - 1.0)
G1_control_pc_anom = 100.0 * (precip_flat[2] / precip_flat[0] - 1.0)
G1_CO2_pc_anom = 100.0 * (precip_flat[2] / precip_flat[1] - 1.0)

X = plt.hist([CO2_control_pc_anom, G1_control_pc_anom], 100, histtype='step', color = ['r','b'], range=[-100,100]);
```
